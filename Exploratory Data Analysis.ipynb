```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
pd.set_option('display.max_columns',None)
df = pd.read_csv('cleaned_salary_data.csv')
df.head()
```

```python
df.describe()
```

```python
sns.set_style('darkgrid')
```

### Distribution plot and Box plot 

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.distplot(df.Rating, kde = True, bins=15, fit=norm)

plt.subplot(1,2,2)
df.boxplot(column = 'Rating')
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.distplot(df['Average Salary'], kde = True, bins=15, fit=norm)

plt.subplot(1,2,2)
df.boxplot(column = 'Average Salary')
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.distplot(df['Company Age'], kde = True, bins=15, fit=norm)

plt.subplot(1,2,2)
df.boxplot(column = 'Company Age')
plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.distplot(df['dcrpn_len'], kde = True, bins=15, fit=norm)

plt.subplot(1,2,2)
df.boxplot(column = 'dcrpn_len')
plt.show()
```

## Heatmaps

```python
corrmat = df[['Rating','hourly','Average Salary','Company Age','python_r',
       'tableau_r', 'excel_r', 'sql_r', 'powerbi_r', 'aws_r', 'spark_r','dcrpn_len', 'Emp_no']].corr()
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corrmat,linewidth=0,vmin=-0.5, square=True)
fig.suptitle('Heatmap')
```

```python
corrmat = df[['Rating','Average Salary','Company Age','dcrpn_len', 'Emp_no']].corr()
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(corrmat,linewidth=1,vmax=0.8,vmin=-0.2 ,square=True, cmap='Blues')
fig.suptitle('Heatmap')
corrmat
```

#### Categorical variables

```python
df.columns
```

```python
df_cat = df[['Location', 'Headquarters','Type of ownership', 'Industry', 'Sector','Revenue','Company',
             'State','isheadquarters','python_r','tableau_r', 'excel_r', 'sql_r','powerbi_r', 'aws_r', 'spark_r','Job_cat', 'Seniority']]
df_cat.head()
```

### Bar plots of categorical variables

```python
for i in df_cat.columns:
    a = df_cat[i].value_counts()
    if len(a) < 61:
        gph = sns.barplot(x=a.index, y=a.values)
        gph.set_xticklabels(gph.get_xticklabels(), rotation = 90)
        plt.title(i)
        plt.show()
```

```python
# display only top 25 since data has too many values
for i in df_cat.columns:
    a = df_cat[i].value_counts()
    if len(a) > 61:
        gph = sns.barplot(x=a.index[:25], y=a.values[:25])
        gph.set_xticklabels(gph.get_xticklabels(), rotation = 90)
        plt.title(i)
        plt.show()
```
